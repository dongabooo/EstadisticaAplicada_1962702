```python
#histograma de la variable dependiente
plt.hist(rossi['week'], bins = 30)
plt.ylim(0,40)
```

```python
#estimar el modelo
cph1 = CoxPHFitter().fit(rossi, 'week', 'arrest')
cph1.print_summary()
```

"fin" -> No rechazamos H0, significa que la variable "fin" no tiene efecto sobre el riesgo de ser arrestado.

"age" -> Rechazamos H0, disminuye el riesgo de ser arrestado.  Por cada año de edad el riesgo de volver a ser arrestado disminuye 6%

"race" -> No rechazamos, ser negro no tiene efecto sobre el riesgo de volver a ser arrestado.

"wexp", "mar", "paro" -> No rechazamos H0; no tienen efecto en el riesgo de volver a ser arrestado"

"prio" -> Rechazamos H0; haber sido arrestado con anterioridad aumenta el riesgo, por cada vez que ha sido arrestado el riesgo de volver a ser arrestado aumenta un 10%.
