```python
#cargar librerías
import pandas as pd
import numpy as np
```

```python
#cargar datos a utilizar
calificaciones = pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/Calificaciones.csv')
calificaciones
```

```python
calificaciones.info()
```

```python
calificaciones.shape
```

```python
calificaciones.head(3)
```

```python

calificaciones.tail(3)
```

```python
calificaciones = calificaciones[['Nombre', 'Apellido', 'Calficación Final']]
calificaciones
#alternativa
#calificaciones2 = calificaciones.drop('Id')
#calificaciones2
```

```python
calificaciones= calificaciones.rename(columns={'Calficación Final' : 'CalFinal'})
calificaciones
```

```python
#promedio
calificaciones['CalFinal'].mean()
```

```python
calificaciones['Situacion'] = np.where(calificaciones['CalFinal'] > 70, 'Aprobado', 'Reprobado')
calificaciones.head()
```

```python
calificaciones.sort_values('CalFinal', ascending=False).head(3)
```

```python
calificaciones.groupby('Situacion')['CalFinal'].count()
```

```python
calificaciones.groupby('Situacion')['CalFinal'].mean().round(1)
```

```python
calificaciones.groupby('Situacion')['CalFinal'].agg(['count', 'mean', 'std']).round(1)
```

## matplotlib

librería para graficar en Python

```python
import matplotlib.pyplot as plt
```

```python
edades = [25,28,22,26]
personas = ['Juan', 'Maria', 'Robert', 'Ana']
```

```python
plt.bar(personas, edades)
plt.show()
```

```python
plt.bar(personas, edades, color = ['blue', 'yellow', 'red', 'green'])
plt.title('edades de 4 personas')
plt.xlabel('personas')
plt.ylabel('edad en años')
plt.show()
```
