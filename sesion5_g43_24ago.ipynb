```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#datos
personas = ['Juan', 'Jorge', 'Ana', 'Alba']
edades = [23,28,18,31]
```

```python
## 2
dic1 = {
    'municipio' : ['Escobedo', 'Apodaca', 'Monterrey', 'Guadalupe'],
    'nacimientos' : [345,487,372,298],
    'muertes' : [25,56,39,67],
    'divorcios' : [28,46,95,19]
}
datos_mun = pd.DataFrame(dic1)
datos_mun
```

```python
##grafica de barras de las edades
plt.bar(personas, edades)
plt.title('edades de 4 desconocidos')
plt.xlabel('persona')
plt.ylabel('edad en años')
plt.show()
```

```python
#grafica de barras para los nacimientos
plt.bar(datos_mun['municipio'], datos_mun['nacimientos'])
plt.title('nacimientos de municipios de NL')
plt.xlabel('municipio')
plt.ylabel('nacimientos')
plt.show()
```

```python
#definir el pizarron/lienzo

fig, (panel1, panel2) = plt.subplots(2,1)
panel1.bar(datos_mun['municipio'], datos_mun['nacimientos'])
panel2.bar(personas, edades)

plt.show()
```
